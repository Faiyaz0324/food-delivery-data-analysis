```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
sns.set(style='whitegrid')
```

```python
df = pd.read_csv(r'/Users/mdfaiyaz/Desktop/food_delivery_data/food_delivery_data.csv')
df
```

```python
print(df.head())
df
```

```python
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.")
```

```python
# Column info
print(df.info())
```

```python
df.rename(columns={'Reached.on.Time_Y.N': 'Reached_On_Time'}, inplace=True)
df
```

```python
print(df.isnull().sum())
```

```python
print("Missing values per column:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

```

```python
# Remove duplicates if any
df = df.drop_duplicates()
```

```python
print(df.dtypes)
```

```python
sns.countplot(data=df, x='Reached_On_Time', palette='Set2')
plt.title('Delivery Timeliness')
plt.xlabel('Delivery Status (0 = On Time, 1 = Late)')
plt.ylabel('Number of Deliveries')
plt.show()
```

```python
sns.countplot(data=df, x='Mode_of_Shipment', palette='pastel')
plt.title('Modes of Shipment')
plt.xlabel('Mode')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(data=df, x='Warehouse_block', palette='muted')
plt.title('Shipments by Warehouse Block')
plt.show()
```

```python
sns.countplot(data=df, x='Product_importance', palette='Set1')
plt.title('Product Importance Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Mode_of_Shipment', hue='Reached_On_Time', data=df, palette='Set2')
plt.title('Shipment Mode vs On-Time Delivery')
plt.xlabel('Mode of Shipment')
plt.ylabel('Count')
plt.legend(title='Reached_On_Time')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(x='Customer_rating', data=df, palette='coolwarm')
plt.title('Customer Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Number of Customers')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Reached_On_Time', y='Discount_offered', data=df)
plt.title('Discount Offered vs Delivery Status')
plt.xlabel('Reached_On_Time (0 = Yes, 1 = No)')
plt.ylabel('Discount Offered')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x='Reached_On_Time', y='Weight_in_gms', data=df)
plt.title('Product Weight vs Delivery Status')
plt.xlabel('Reached_On_Time')
plt.ylabel('Weight in Grams')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.countplot(x='Product_importance', hue='Gender', data=df, palette='pastel')
plt.title('Product Importance by Gender')
plt.xlabel('Product Importance')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Reached_On_Time', y='Customer_care_calls', data=df)
plt.title('Customer Care Calls vs Delivery')
plt.xlabel('Reached_On_Time')
plt.ylabel('Customer Care Calls')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Prior_purchases', y='Cost_of_the_Product', hue='Reached_On_Time', data=df)
plt.title('Prior Purchases vs Product Cost')
plt.xlabel('Prior Purchases')
plt.ylabel('Cost of the Product')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Product_importance', y='Discount_offered', data=df, palette='muted')
plt.title('Discounts based on Product Importance')
plt.xlabel('Product Importance')
plt.ylabel('Discount Offered')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Warehouse_block', hue='Reached_On_Time', data=df, palette='husl')
plt.title('Warehouse Block vs Delivery Status')
plt.xlabel('Warehouse Block')
plt.ylabel('Count')
plt.legend(title='Reached_On_Time')
plt.show()
```

```python
plt.figure(figsize=(8,5))
delivery_rate = df.groupby('Product_importance')['Reached_On_Time'].value_counts(normalize=True).unstack().fillna(0)
delivery_rate.plot(kind='bar', stacked=True, colormap='coolwarm', figsize=(8,5))
plt.title('Delivery Success Rate by Product Importance')
plt.ylabel('Proportion')
plt.xlabel('Product Importance')
plt.legend(['On Time', 'Late'], title='Delivery Status')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.pairplot(df, vars=['Cost_of_the_Product', 'Discount_offered', 'Weight_in_gms'], 
             hue='Reached_On_Time', palette='Set2', plot_kws={'alpha':0.6})
plt.suptitle('Pairwise Feature Analysis', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(7,5))
avg_discount = df.groupby('Mode_of_Shipment')['Discount_offered'].mean().sort_values()
avg_discount.plot(kind='barh', color='teal')
plt.title('Average Discount by Shipment Mode')
plt.xlabel('Average Discount')
plt.ylabel('Mode of Shipment')
plt.show()
```

```python
late_ratio = df[df['Reached_On_Time'] == 1]['Warehouse_block'].value_counts(normalize=True)
late_ratio.plot(kind='bar', color='crimson', figsize=(7,5))
plt.title('Proportion of Late Deliveries by Warehouse Block')
plt.ylabel('Percentage of Late Deliveries')
plt.xlabel('Warehouse Block')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Customer_rating', hue='Reached_On_Time', palette='Set1')
plt.title('Delivery Timeliness by Customer Rating')
plt.xlabel('Customer Rating')
plt.ylabel('Number of Orders')
plt.legend(['Late', 'On Time'])
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Cost_of_the_Product', y='Discount_offered', hue='Reached_On_Time', alpha=0.7)
plt.title('Discount Offered vs Product Cost')
plt.xlabel('Cost of Product')
plt.ylabel('Discount Offered')
plt.legend(['Late', 'On Time'])
plt.show()
```

```python
gender_late = df[df['Reached_On_Time'] == 1]['Gender'].value_counts()
plt.figure(figsize=(6,6))
gender_late.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Late Deliveries by Gender')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.countplot(data=df, x='Product_importance', palette='Set2')
plt.title('Distribution of Product Importance')
plt.xlabel('Product Importance')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(data=df, x='Reached_On_Time', y='Customer_care_calls', palette='coolwarm')
plt.title('Customer Care Calls vs Delivery Status')
plt.xticks([0, 1], ['On Time', 'Late'])
plt.ylabel('Number of Calls')
plt.show()
```
